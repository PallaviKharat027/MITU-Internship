## Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([23, 34, 42, 49, 55, 61, 68, 73, 78, 82])
y = np.array([25, 29, 33, 37, 44, 47, 57, 62, 67, 71])
z = np.array([87, 82, 78, 73, 67, 63, 55, 51, 47, 39])
```

### Line Plot

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 5);
plt.grid()

```

```python
# marlers => >, <, ^, H, o, * 

plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = '^', mfc = 'black', color = 'pink', ls = '--',
        ms = 5);
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, y, marker = 'o', mfc = 'red', color = 'black', ls = '--',
        ms = 5, label= '2020');
plt.plot(x, z, marker = 'o', mfc = 'yellow', color = 'black', ls = '--',
        ms = 5, label= '2030');
plt.grid()
plt.legend()
```

#### Sub Plot

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'black', color = 'black', ls = '--',
        ms = 5);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1, 90, 5))
plt.yticks(range(1, 90, 5))
plt.plot(x, z, marker = 'o', mfc = 'black', color = 'black', ls = '--',
        ms = 5);
plt.grid()
```

### Bar Plot

```python
courses = ['java', 'python', 'kotlin', 'dart', 'R']
students = [67, 63, 59, 55, 47]
```

```python
plt.bar(courses, students, color = 'magenta', edgecolor = 'black', width = 0.3,
        linewidth = 3, alpha = 0.5);
plt.title('Student Enrollment')
plt.xlabel('courses')
plt.ylabel('students')
plt.grid()
```

## Box Plot

```python
marks = [67, 65, 55, 59, 69, 72, 51, 70, 53, 98,23]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks, 75)
q1 = np.percentile(marks, 25)
q3, q1
```

```python
iqr = q3 - q1
iqr
```

```python
ub = q3 + (1.5 * iqr)
lb = q1 - (1.5 * iqr)
```

```python
ub # we get outliens
```
